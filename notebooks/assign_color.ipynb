```python
import os
import yaml
import numpy as np

data = [x.strip() for x in open('data_rlbench/urdf/basketball_in_hoop/basket_ball_hoop_visual.ply').readlines()]
```

```python
cfg_idx = 0
while True:
    cfg_path = f'data_rlbench/urdf/basketball_in_hoop/basket_ball_hoop_visual_{cfg_idx}.yaml'
    if not os.path.exists(cfg_path):
        break
    cfg = yaml.load(open(cfg_path), Loader=yaml.FullLoader)

    indices = np.array(cfg['indices']).reshape(-1, 3)
    colors = (np.array(cfg['colors']).reshape(-1, 3) * 255).astype(np.uint8)

    for i, line in enumerate(data):
        if line.startswith('3 '):
            face_vs = np.array(line.removeprefix('3 ').split(' '), dtype=np.int32)
            if not face_vs.shape == (3,):
                continue
            face_vs_idx = np.where(np.all(indices == face_vs[None, :], axis=-1))[0]
            if len(face_vs_idx) == 0:
                continue
            else:
                data[i] = line + ' ' + ' '.join([str(x) for x in colors[0].tolist()])
    
    cfg_idx += 1
    
    # break

## Remaining faces
colors = np.array([[255, 255, 255]])
for i, line in enumerate(data):
    if line.startswith('3 '):
        print(data[i])
        face_vs = np.array(line.removeprefix('3 ').split(' '), dtype=np.int32)
        if face_vs[0] == 5392:
            print(face_vs)
            print(face_vs.shape)
        # break
        if face_vs.shape == (3,):
            data[i] = line + ' ' + ' '.join([str(x) for x in colors[0].tolist()])
        print(data[i])
```

```python
for line in data:
    if line.startswith('3 '):
        print(line)
        break 
```

```python
end_header_idx = data.index('end_header')   
data.insert(end_header_idx, 'property uchar red')
data.insert(end_header_idx + 1, 'property uchar green')
data.insert(end_header_idx + 2, 'property uchar blue')

with open('data_rlbench/urdf/basketball_in_hoop/basket_ball_hoop_visual_colored.ply', 'w') as f:
    f.write('\n'.join(data))
```
